```python
# Preanalisis jc
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df1 = pd.read_csv("df_final_web_data_pt_1.txt")
df2 = pd.read_csv("df_final_web_data_pt_2.txt")

df = pd.concat([df1, df2], ignore_index=True)

```

```python
# Conversión de date_time a formato datetime

df["date_time"] = pd.to_datetime(df["date_time"], errors="coerce")
df["date_time"].isna().sum(), df["date_time"].min(), df["date_time"].max()
```

```python
# Normalización de etiquetas de process_step

df["process_step"] = df["process_step"].astype(str).str.strip().str.lower()
df["process_step"].value_counts()
```
